```python
import nltk
# nltk.download("punkt")
from nltk.stem.lancaster import LancasterStemmer
stemmer = LancasterStemmer()
```

```python
import tensorflow as tf
import numpy as np
import tflearn
import random
import json
```

```python
with open("dataset.json") as file:
    dataset = json.load(file)
```

```python
words = []
ignore = ['{', '}', "'Questions", "'Answers", ':', "'" ,"''", ',', '?', '!', ]
```

```python
for intent in dataset['intents']:
    for questions in intent['Questions']:
        print(questions)
        w = nltk.word_tokenize(questions)
        words.extend(w)
          

# words = [stemmer.stem(w.lower()) for w in words if w not in ignore]
# for w in words:
#     if w in ignore:
#         words.remove(w)
```

```python
training = []
output = []

output_empty = [0] * len(words)
```

```python
for w in words:
    bag = []
    pattern_word = w[0]
    
    pattern_word = [stemmer.stem(word.lower()) for word in pattern_word]
    bag.append(1) if w in pattern_word else bag.append(0)
    
    output_row = list(output_empty)
    training.append([bag, output_row])
    
```
